```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x=pd.read_csv("C:\\Users\\Vijay\\Downloads\\oyorooms.csv")
y=x.head(10)
data=np.array(y)
j=0
for i in range(10):
        plt.bar(j,data[i][4][1:],label="example2",color="y")
        plt.bar(j+1,data[i][5][1:],label="example3",color="k")
        j+=2
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('oyodetails')
print.show()
```
