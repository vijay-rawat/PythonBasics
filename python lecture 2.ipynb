# stack plot

```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5]
sleeping=[7,5,4,2,7]
eating=[2,4,6,8,1]
working[2,4,5,6,8]

plt.plot([],color='m',label='sleeping',linewidth=5)
plt.plot([],color='k',label='eating',linewidth=5)
plt.plot([],color='r',label='working',linewidth=5)

plt.stackplot(days,sleeping,working,eating)
colours=(['m','k','r'])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#pie chart
```

```python
import matplotlib.pyplot as plt
slices=[7,2,2,12]
activities=['sleeping','eating','working','playing']
cols=['c','m','r','b']
plt.pie(slices,labels=activities,colors=cols,startangle=90,shadow=True, explode=(0,0.1,0,0),autopct='%1.1%%')
plt.title('pie plot')
plt.show()
```
